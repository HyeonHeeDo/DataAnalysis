# 머신러닝 시스템 종류
## 1. 지도학습 / 비지도학습 / 준지도학습 + 강화학습  
### 지도학습 Supervised Learning
* 알고리즘에 주입하는 훈련 데이터에 원하는 답 __레이블 label__ 포함  
  
  
* 지도 학습 작업
    * 분류 classification
    * 회귀 regression : 예측 변수 (특성 feature) 이용한 타깃 수치 예측
  
  
* 지도 학습 알고리즘
    * k-최근접 이웃
    * 선형회귀
    * 로지스틱 회귀
    * 서포트 벡터 머신
    * 결정 트리 & 랜덤 포레스트
    * 신경망
    
### 비지도학습 Unsupervised Learning
* 훈련 데이터에 레이블 없음  
* 이상치 탐지
  
  
* 비지도 학습 알고리즘
    * 군집 clustering
        * k-평균
        * 계층 군집 분석
        * 기댓값 최대화
    * 시각화와 차원축소 - 특성 추출
        * 주성분 분석
        * 커널 PCA
        * 지역적 선형 임베딩
        * t-SNE
    * 연관 규칙 학습
        * 어프라이어리
        * 이클렛
        
### 준지도학습 Semisupervised Learning
* 일부 데이터는 레이블 존재, 레이블 없는 데이터 다수 일 때  
  
  
* 준지도 학습 알고리즘 = 지도 학습 + 비지도 학습 조합

### 강화학습 Reinforcement Learning
* __에이전트__가 __환경__을 관찰하여 행동한 결과에 대한 __보상/벌점__을 바탕으로 정책 (최선의 전략)을 스스로 학습

***

## 2. 배치학습 / 온라인 학습
: 입력 데이터의 스트림stream으로부터 점진적 학습이 가능한가?
### 배치학습 (오프라인 학습)
* 점진적 학습 불가능
  
  
* 가용데이터를 모두 사용
* 새로운 데이터 입력시 기존 데이터와 함께 전체 데이터로 다시 훈련


### 온라인학습
* 미니 배치 단위로 훈련
  
  
* 파라미터 : 학습률 learning rate

***

## 3. 사례 기반 / 모델 기반
: 일반화 할 때

### 사례 기반
* 샘플을 기억한 후, 새로운 샘플에 대해 기존 샘플과의 유사도 측정
  
### 모델 기반
* 모델 생성 후 예측
    * 효용함수, 적합도함수, 비용함수 이용해서 평가

# 나쁜 데이터
#### 1. 충분치 않은 훈련 데이터


#### 2. 대표성이 없는 훈련 데이터
* 샘플이 적으면 : 샘플링 잡음
* 샘플이 많으나 표본 추출 방법이 잘못되었으면 : 샘플링 편향


#### 3. 낮은 품질의 데이터 
* 에러, 이상치, 잡음


#### 4. 관련 없는 특성 포함
* 특성공학 이용해 좋은 특성 찾을 것
     * 특성 선택 - 특성 추출

# 나쁜 알고리즘

#### 1. 과대적합
: 훈련 데이터에는 잘 맞지만 일반성이 떨어지는 경우 (예측을 믿기 힘듦)

* 파라미터 수 적은 모델 선택 / 훈련 데이터 특성 줄이기 / 모델을 규제하여 단순화
  
  
* 더 많은 훈련 데이터 이용


* 훈련 데이터 잡음 감소

> 규제
* 모델에 제약을 줘서 단순하게 하는 것
* <u>하이퍼파라미터</u>가 결정
    * 학습하는 동안 적용할 규제의 양
    * 학습 알고리즘의 파라미터 (모델 파라미터 X)
    * 학습 알고리즘으로부터 영향을 받지 않음
    * 훈련 전에 미리 지정
    * 훈련 중에는 고정된 상수값

#### 2. 과소적합
: 모델이 단순해서 데이터의 구조를 학습하지 못하는 경우
  
  
* 모델 파라미터가 많은 강력한 모델을 선택


* 학습 알고리즘에 더 좋은 특성을 제공 (특성공학)


* 모델 제약을 줄인다
    * 규제에서 하이퍼파라미터의 수 줄이기

# 모델 평가
#### 1. 훈련 세트
: 모델 훈련용 

#### 2. 테스트 세트
: 모델 평가용
* <u>일반화 오차</u> 측정값
    * 새로운 샘플에 대한 오차 비율
    * 새로운 샘플에 모델이 얼마나 잘 작동할지 판단


* 훈련오차가 작고 일반화 오차가 크면 과대적합 된 것!


#### 일반화 오차를 테스트 세트에서 여러번 측정하면
모델과 하이퍼파라미터가 테스트 세트에 최적화된 모델 생성  
= 새로운 데이터가 잘 작동하지 않을 수 있음 !


#### 3. 검증 세트
* 훈련 세트 : 다양한 하이퍼파라미터로 여러 모델 훈련

* 검증 세트 : 최상의 성능을 나타내는 하이퍼파라미터 & 모델 선택
    * 교차 검증

* 테스트 세트 : 일반화 오차의 추정값 구함
    * 단 한번만 수행
    

> 교차 검증
* 훈련 데이터에서 너무 많은 데이터가 검증 세트로 넘어가지 않도록 한다.
    1. 훈련세트를 여러 서브셋subset으로 나누기
    2. 각 모델을 서브셋 조합별로 훈련, 나머지 부분으로 검증 수행
    3. 모델과 하이퍼파라미터를 선택
    4. 선택한 하이퍼파라미터를 전체 훈련 데이터에 적용하여 최종 모델 훈련
    5. 테스트 세트에서 일반화 오차 측정
